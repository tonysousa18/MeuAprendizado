# **LISTAS CIRCULARES**

## Em uma lista encadeada circular, as extremidades se conectam, o que significa que o ultimo nó da lista aponta para o primeiro nó. E

## Em uma lista simplesmente encadeada circular, o ultimo nó aponta para o primeiro nó.

## Em uma lista duplamente encadeada circular, o primeiro nó aponta para o ultimo nó e ultimo nó aponta de volta para o primeiro.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class ListaCircular:
    def __init__(self):
        self.tail = None
        self.head = None
        self.size = 0

    def iter(self):
        current = self.head

        while current:
            val = current.data
            current = current.next
            yield val

    def contar_elementos(self):
        if not self.head:
            return 0
        
        counter = 0
        current = self.head
        
        while current.next != self.head:
            counter += 1
            current = current.next

        return counter

            
    def append(self, data):
        new_node = Node(data)

        if self.tail: # Se já haver elemento no fim da lista:
            self.tail.next = new_node #O proximo elemento do fim da fila sera o novo Nó
            self.tail = new_node #O novo Nó final é o novo nó
            new_node.next = self.head #O novo nó apontará para head

        else: #Se nao haver nada na lista
            self.head = new_node #O primeiro nó sera o novo nó
            self.tail = new_node #O ultimo nó sera o novo nó
            self.tail.next = self.tail #O ultimo nó apontará para si mesmo

        self.size += 1

    def add_node_after_existing_node(self, data, target_node):
        new_node = Node(data)
        node = self.head # Começamos a busca a partir da cabeça da lista.
        find = False    

        #Percorremos a lista procurando pelo nó que tem o valor target_node.
        while node != None:
            if node.data == target_node:
                new_node.next = node.next #Quando encontramos, fazemos new_node.next = node.next para encaixá-lo na posição correta.
                node.next = new_node
                find = True
                break

            node = node.next

        if not find:
            print(f"Target node{target_node} not found")
            #Se não acharmos o nó alvo, mostramos uma mensagem de erro


    def delete(self, data):
        current = self.head
        prev = self.head
        flag = False

        while prev == current or prev != self.tail: #Enquanto o Nó anterior for igual ao nó atual ou o nó anterior for diferente do nó final
            if current.data == data: #Se o nó atual for igual ao dado buscado
                if current == self.head: #O item a ser escolhido é o nó head
                    self.head = current.next #O head vai ser o nó proximo a ele
                    self.tail.next = self.head #O nó proximo ao tail é o novo head
                    flag = True #Exclusao completa

                elif current == self.tail:#O item a ser excluido é o nó tail
                    self.tail = prev #O tail atual agora é o anterior
                    prev.next = self.head #O nó que agora é o tail aponta pro head
                    flag = True #Exclusao completa

                else:
                    #O item a ser excluido é o no intermediario
                    prev.next = current.next #O nó anterior agora aponta pro proximo nó do nó escolhido
                    self.size -= 1 #
                    flag = True

                return
            
            prev = current
            current = current.next
        if flag is False:
            print('O item que voce quer excluir nao esta presente na lista')


```

```python
lista_circular = ListaCircular()
print('Adicionando Dadors:')
lista_circular.append('Gabinete')
lista_circular.append('Placa mae')
lista_circular.append('Memoria Ram')
lista_circular.append('Processador')

counter = 0
for elemento in lista_circular.iter():
    print(elemento)
    counter += 1

    if counter > lista_circular.contar_elementos():
        break

print('\n Adicionando em local: ')
lista_circular.add_node_after_existing_node('Placa pai', 'Placa mae')

counter = 0
for elemento in lista_circular.iter():
    print(elemento)
    counter += 1

    if counter > lista_circular.contar_elementos():
        break



print('\n Apagando dadors: ')
lista_circular.delete('Placa pai')

counter = 0
for elemento in lista_circular.iter():
    print(elemento)
    counter += 1

    if counter > lista_circular.contar_elementos():
        break
```
