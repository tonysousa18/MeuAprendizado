# **LISTAS ENCADEADAS** 

### Uma lista encadeada é uma estrutura de dados na qual os elementos de dados são armazenados em uma ordem linear, por meio de estruturas baseadas em ponteiros. Eles armazenam os dados e o local, e o local armazena a posicao do proximo item de dados na memória.

## **Arrays**

#### Arrays são colecoes de itens de dados do mesmo tipo, enquanto listas encadeadas é um array armazenado sequencialmente e conectados por ponteiros. Em listas os itens são armazenados em espacos diferentes da memoria, ja no array os elementos sao armazenados em locais contiguos

#### É muito rápido armazenar, pecorrer e acessar elementos de um array em comparação com listas encadeadas pois os elementos podem ser acessados aleatoriamente com uso das posicoes indicadas pelo seu indice, enquanto, no caso de uma lista encadeada, eles sao acessados sequencialmente.

#### Portando, arrays são adequados quando você quiser executar operacoes de acesso a elementos e poucas operações de insercao e exclusao, enquanto as listas encadeadas sao apropriadas em aplicacoes nas quais o tamanho da lista nao é fixo e muitas operacoes de insercao e exclusao sao necessarias.

# **Introdução a listas encadeadas**

#### A lista encadeada tem as seguintes propriedades: 
##### 1. Os elementos de dados são armazenados na memória em locais diferentes conectados por ponteiros. Um ponteiro é um objeto que pode armazenar o endereco de memória de uma variável, e cada elemento aponta para o próximo elemento de dados até que seja alcancado o ultimo elemento, que aponta pra None.
##### 2. O tamanho da lista pode aumentar ou diminuir durante a execução do programa.

#### Ao contrário dos Arrays, a lista encadeada armazena os itens de dados sequencialmente em locais diferentes na memória, onde cada item de dado é armazenado separadamente e vinculado aos outros itens de dados com o uso de ponteiros. Cada um desses dados se chama Nó, e um Nó armazena dados reais (Data) e um ponteiro (Next).

## O nó "pão" aponta para "manteiga", e "manteiga" aponta para "café", mas café não aponta para ninguem, ou seja: **None**
## Primeiro, conheceremos a implementação do nó:

```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

```

##### Aqui, o ponteiro next é inicializado com None, ou seja, a menos que alteremos o valor de **next**, o nó será um endpoint, ou seja, qualquer nó anexado a lista será independente

## Existem 3 tipos de lista - simplesmente encadeada, duplamente encadeada, e encadeada circular. Comecaremos discutindo as listas simplesmente encadeadas.
## Precisamos conhecer as operações a seguir para usar uma lista encadeada para aplicacoes de tempo real:

- Pecorrer a lista
- Inserir um item de dados na lista:
    - Inserir um novo item de dados(nó) no começo
    - Inserir um novo item de dados no fim
    - Inserir um novo item de dados(nó) no meio ou qualquer posição

- Excluir um item da lista:
    - Excluir o primeiro nó
    - Excluir o ultimo nó
    - Excluir um nó do meio ou qualquer posição




# **Listas simplesmente encadeadas**
### A lista simplesmente encadeada contem varios nós e cada nó contem dados e um ponteiro que aponta pro proximo nó.

# **Criando e pecorrendo uma lista simplesmente encadeada:**

```python
# Criando os dados para cada nó
n1 = Node('pão')
n2 = Node('manteiga')
n3 = Node('café')

# Criando os vinculos entre os nós
n1.next = n2
n2.next = n3
```

# **Pecorrendo a lista simplesmente encadeada**
## Para pecorrer, comecaremos com uma variavel current com o primeiro nó, e depois pecorreremos a lista completa com um loop, visitando cada nó:

```python
current = n1

while current:
    print(current.data)
    current = current.next
```

## No loop, exibimos o elemento atual, e depois configuramos current para apontar para o proximo elemento d lista.
## Essa é a maneira mais simples de pecorre-la, porem há uma maneira mais eficiente

```python
class Node:
    def __init__(self, data):
        self.data = data #Armazena o valor do nó (por exemplo, "banana", "123", etc.).
        self.next = None # Inicialmente, o próximo nó é None (ou seja, esse nó não aponta para nenhum outro nó ainda).


class LinkedList: 
    def __init__(self):
        self.head = None # Inicialmente, a lista começa vazia, então head (o primeiro nó da lista) é None.
    

    def add_node_at_start(self, data):
        new_node = Node(data) # Criamos um novo nó com o valor data.
        new_node.next = self.head # O novo nó passa a apontar para o antigo primeiro nó.
        self.head = new_node # Agora, o novo nó vira a cabeça da lista.

    def add_node_at_end(self, data) -> None:
        new_node = Node(data) #Criamos um novo nó com o valor data.

        # Se a lista está vazia, basta colocar esse nó como cabeça (head).
        if self.head == None:
            new_node.next = self.head
            self.head = new_node 

        else:
            #Caso contrário, percorremos a lista até encontrar o último nó (while previous_node.next != None:)
            previous_node = self.head
            while previous_node.next != None:
                previous_node = previous_node.next

            previous_node.next = new_node #  Ligamos o último nó ao novo nó.
            new_node.next = None

    def add_node_after_existing_node(self, data, target_node):
        new_node = Node(data)
        node = self.head # Começamos a busca a partir da cabeça da lista.
        find = False    

        #Percorremos a lista procurando pelo nó que tem o valor target_node.
        while node != None:
            if node.data == target_node:
                new_node.next = node.next #Quando encontramos, fazemos new_node.next = node.next para encaixá-lo na posição correta.
                node.next = new_node
                find = True
                break

            node = node.next

        if not find:
            print(f"Target node{target_node} not found")
            #Se não acharmos o nó alvo, mostramos uma mensagem de erro

    def remove_node(self, data) -> None:
        node = self.head
        find = False

        if node.data == data:
            self.head == node.next # Se o nó a ser removido for a cabeça, basta mover self.head para o próximo nó.

        else:
            #Se não for, percorremos a lista até achar o nó antes do que queremos excluir (previous_node).
            previous_node = node
            node = node.next #

        while node != None:
            if node.data == data:
                previous_node.next = node.next # Pulamos o nó a ser removido
                find = True
                break

            else:
                previous_node = node
                node = node.next

        if not find:
            print('o nó nao foi encontrado')

    def __str__(self) -> str:
        list_nodes = []
        node = self.head

        while node is not None:
            list_nodes.append(node.data)
            node = node.next

        list_nodes.append("None")

        return " --> ".join(list_nodes)
```
